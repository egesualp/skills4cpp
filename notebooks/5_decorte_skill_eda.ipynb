```python
df = decorte.merge(esco, left_on='esco_id', right_on='occupationUri', how='left')
```

```python
# Group by occupationUri and count skills at each level
skill_counts_per_occupation = df.groupby('occupationUri').agg({
    'level0_label': 'nunique',
    'level1_label': 'nunique', 
    'level2_label': 'nunique',
    'level3_label': 'nunique'
}).rename(columns={
    'level0_label': 'level0_skill_count',
    'level1_label': 'level1_skill_count',
    'level2_label': 'level2_skill_count', 
    'level3_label': 'level3_skill_count'
})

# Calculate statistics for each level
print("Statistics for number of skills per occupation at each level:")
print("=" * 60)

for level in ['level0', 'level1', 'level2', 'level3']:
    col_name = f'{level}_skill_count'
    stats = skill_counts_per_occupation[col_name].describe()
    print(f"\n{level.upper()} Skills per Occupation:")
    print(f"Mean: {stats['mean']:.2f}")
    print(f"Median: {stats['50%']:.2f}")
    print(f"Min: {stats['min']:.0f}")
    print(f"Max: {stats['max']:.0f}")
    print(f"Std: {stats['std']:.2f}")
    print(f"25th percentile: {stats['25%']:.2f}")
    print(f"75th percentile: {stats['75%']:.2f}")
    print(f"Available # of categories: {df[f'{level}_label'].nunique()}")

# Display the skill counts dataframe
print(f"\nSkill counts per occupation (showing first 10 rows):")
print(skill_counts_per_occupation.head(10))
```

```python
# Group by occupationUri and count appearances of each level
level_appearances_per_occupation = df.groupby('occupationUri').agg({
    'level0_label': 'count',
    'level1_label': 'count', 
    'level2_label': 'count',
    'level3_label': 'count'
}).rename(columns={
    'level0_label': 'level0_appearances',
    'level1_label': 'level1_appearances',
    'level2_label': 'level2_appearances', 
    'level3_label': 'level3_appearances'
})

# Calculate statistics for level appearances
print("Statistics for number of level appearances per occupation:")
print("=" * 60)

for level in ['level0', 'level1', 'level2', 'level3']:
    col_name = f'{level}_appearances'
    stats = level_appearances_per_occupation[col_name].describe()
    print(f"\n{level.upper()} Appearances per Occupation:")
    print(f"Mean: {stats['mean']:.2f}")
    print(f"Median: {stats['50%']:.2f}")
    print(f"Min: {stats['min']:.0f}")
    print(f"Max: {stats['max']:.0f}")
    print(f"Std: {stats['std']:.2f}")
    print(f"25th percentile: {stats['25%']:.2f}")
    print(f"75th percentile: {stats['75%']:.2f}")

# Display the level appearances dataframe
print(f"\nLevel appearances per occupation (showing first 10 rows):")
print(level_appearances_per_occupation.head(10))

# Show overall averages
print(f"\nOverall averages:")
for level in ['level0', 'level1', 'level2', 'level3']:
    col_name = f'{level}_appearances'
    avg = level_appearances_per_occupation[col_name].mean()
    print(f"On average, 1 occupation has {avg:.2f} {level} appearances")
```
