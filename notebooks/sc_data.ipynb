```python
import pandas as pd
import numpy as np
import os
```

```python
# iterate over all files in data/title_pairs, save dataframes in a dictionary
title_pairs = {}
for file in os.listdir('../data/title_pairs'):
    title_pairs[file] = pd.read_csv(f'../data/title_pairs/{file}')
```

```python
# print current path
title_pairs.keys()
```

```python
# print column names of the dataframes
for key in title_pairs.keys():
    print(key)
    print(title_pairs[key].columns)
    print("--------------------------------")
```

```python
# concat all dataframes in title_pairs, add column with filename = key
for key in title_pairs.keys():
    title_pairs[key]['filename'] = key

all_pairs = pd.concat(title_pairs.values(), ignore_index=True)
```

```python
pool = pd.read_csv('../data/occupations_en.csv')
all_pairs = all_pairs.merge(pool[['conceptUri', 'preferredLabel']], left_on='esco_id', right_on='conceptUri', how='left')
all_pairs.head()
```

```python
# check conceptUri coverage by filename
coverage = all_pairs.groupby('filename')['conceptUri'].apply(lambda x: (x.notna().sum() / len(x)) * 100)
print("Coverage of conceptUri by filename (%):")
print(coverage)
```

```python
all_pairs.groupby('filename').count()
```

```python
# for each dataframe in title_pairs, map pool with conceptUri column
for key in title_pairs.keys():
    title_pairs[key] = title_pairs[key].merge(pool, left_on='esco_id', right_on='conceptUri', how='left')
```
