```python
import matplotlib.pyplot as plt
# Distribution of skills based on how many occupations have each skill
skill_occupation_counts = df_final_1.groupby('skillUri').size().reset_index(name='num_occupations')

# Display statistics
print("Distribution of skills by number of occupations:")
print(f"Min occupations per skill: {skill_occupation_counts['num_occupations'].min()}")
print(f"Max occupations per skill: {skill_occupation_counts['num_occupations'].max()}")
print(f"Mean occupations per skill: {skill_occupation_counts['num_occupations'].mean():.2f}")
print(f"Median occupations per skill: {skill_occupation_counts['num_occupations'].median():.2f}")
print(f"Standard Deviation: {skill_occupation_counts['num_occupations'].std():.2f}")

# Show distribution
print("\nDistribution breakdown:")
print(skill_occupation_counts['num_occupations'].value_counts().sort_index().head(20))

# Plot the distribution
plt.figure(figsize=(12, 6))
plt.hist(skill_occupation_counts['num_occupations'], bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('Number of Occupations')
plt.ylabel('Number of Skills')
plt.title('Distribution of Skills by Number of Occupations')
plt.grid(True, alpha=0.3)
plt.show()

# Show skills that appear in the most occupations
print("\nTop 10 skills by number of occupations:")
top_skills = skill_occupation_counts.nlargest(10, 'num_occupations')
for _, row in top_skills.iterrows():
    skill_label = df_final_1[df_final_1['skillUri'] == row['skillUri']]['skillLabel'].iloc[0]
    print(f"{skill_label}: {row['num_occupations']} occupations")
```

```python
skill_occupation_counts.sort_values('num_occupations', ascending=True, inplace=True)
```

```python
skill_occupation_counts['cumsum'] = skill_occupation_counts['num_occupations'].cumsum()
skill_occupation_counts['cumsum_percentage'] = skill_occupation_counts['cumsum'] / skill_occupation_counts['num_occupations'].sum()
skill_occupation_counts.head()
skill_occupation_counts.query('cumsum_percentage < 0.95')
skill_occupation_counts.query('cumsum_percentage < 0.95').num_occupations.sum()
skill_occupation_counts.query('cumsum_percentage < 0.95').num_occupations.sum() / skill_occupation_counts.num_occupations.sum()
```

```python
# Find a sweetspot breakpoint for num_occupations
# Let's analyze the distribution to find where it becomes "normal"
import numpy as np
# Look at percentiles to understand the distribution
percentiles = [50, 75, 80, 85, 90, 95, 99]
print("Percentiles of num_occupations:")
for p in percentiles:
    value = np.percentile(skill_occupation_counts['num_occupations'], p)
    print(f"{p}th percentile: {value:.1f}")

print(f"\nTotal skills: {len(skill_occupation_counts)}")

# Find potential breakpoints
# Option 1: Skills that appear in <= 10 occupations (likely covers most specific skills)
breakpoint_10 = skill_occupation_counts[skill_occupation_counts['num_occupations'] <= 10]
print(f"\nSkills with <= 10 occupations: {len(breakpoint_10)} ({len(breakpoint_10)/len(skill_occupation_counts)*100:.1f}%)")

# Option 2: Skills that appear in <= 20 occupations
breakpoint_20 = skill_occupation_counts[skill_occupation_counts['num_occupations'] <= 20]
print(f"Skills with <= 20 occupations: {len(breakpoint_20)} ({len(breakpoint_20)/len(skill_occupation_counts)*100:.1f}%)")

# Option 3: Skills that appear in <= 50 occupations
breakpoint_50 = skill_occupation_counts[skill_occupation_counts['num_occupations'] <= 50]
print(f"Skills with <= 50 occupations: {len(breakpoint_50)} ({len(breakpoint_50)/len(skill_occupation_counts)*100:.1f}%)")

# Show the distribution around potential breakpoints
print(f"\nDistribution around potential breakpoints:")
print("Occupations 1-10:", len(skill_occupation_counts[skill_occupation_counts['num_occupations'].between(1, 10)]))
print("Occupations 11-20:", len(skill_occupation_counts[skill_occupation_counts['num_occupations'].between(11, 20)]))
print("Occupations 21-50:", len(skill_occupation_counts[skill_occupation_counts['num_occupations'].between(21, 50)]))
print("Occupations 51-100:", len(skill_occupation_counts[skill_occupation_counts['num_occupations'].between(51, 100)]))
print("Occupations 100+:", len(skill_occupation_counts[skill_occupation_counts['num_occupations'] > 100]))

skill_occupation_counts
```
